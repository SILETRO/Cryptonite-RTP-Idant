# Quantification

```python
date_reg = (date_time - date_time.min()).dt.days
```

```python
date_arr = np.array(date_reg).reshape(-1,1)
df_temp_arr = np.array(df['temperature_2m_mean (°C)']).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(date_arr,df_temp_arr)
```

```python
slope_temp = lin_reg.coef_[0][0]
intercept = lin_reg.intercept_[0]
```

```python
slope_temp*365
```

```python
trend_line_temp = lin_reg.predict(date_arr)
```

```python
total_days = date_arr.max()
change = slope_temp*total_days
change
```

```python
plt.figure(figsize=(12, 6))
plt.plot(date_time, df_temp_arr, label='Daily Data', color='gray', alpha=0.5)
plt.plot(date_time, trend_line_temp, label=f'Trend Line', color='red', linewidth=2)
plt.title(f'Climate Change Impact on temperature (2011-2026)')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
df_prec_arr = np.array(df['precipitation_sum (mm)']).reshape(-1,1)
```

```python
lin_reg.fit(date_arr,df_prec_arr)
```

```python
slope = lin_reg.coef_[0][0]
intercept = lin_reg.intercept_[0]
trend_line_prec = lin_reg.predict(date_arr)
```

```python
change_prec = slope * total_days
change_prec
```
